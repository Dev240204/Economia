```python
import matplotlib.pyplot as plt
```

```python
x = df["Date"]
y = df["Discount"]
  
plt.bar(x, y) 
plt.xlabel("X-axis")  # add X-axis label 
plt.ylabel("Y-axis")  # add Y-axis label 
plt.title("Any suitable title")  # add title 
plt.show()
```

```python
# Using Numpy to create an array X 
X = df["Date"]
  
# Assign variables to the y axis part of the curve 
y = df["Original_price"]
z = df["Discounted_price"]
  
# Plotting both the curves simultaneously 
plt.plot(X, y, color='r', label='Original_price') 
plt.plot(X, z, color='g', label='Discounted_price') 
  
# Naming the x-axis, y-axis and the whole graph 
plt.xlabel("Date") 
plt.ylabel("Price")
plt.xticks(rotation=45)
  
# Adding legend, which helps us recognize the curve according to it's color 
plt.legend() 
  
# To load the display window 
plt.show() 
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
df['Original_price'].array.reshape(-1,1)
```

```python
df['Discounted_price'].values.reshape(1,-1)
```

```python
X=df['Original_price'].array.reshape(-1,1)
X=X[:int(len(df)-5)]
# print(X)
Y=df['Discounted_price'].array.reshape(-1,1)
Y=Y[:int(len(df)-5)]
# print(Y)
```

```python
model.fit(X,Y)
```

```python
#Test 
predictions=model.predict(X)
print("The model score is:", model.score(X,Y))
```

```python
new_data=df[['Original_price']].tail(1)
prediction=model.predict(new_data)
print("The model predicts the last row or day to be:",prediction)
```
